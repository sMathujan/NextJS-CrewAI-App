1. Create the backend foleder.
`mkdir crewai_be`

2. Create project files.

`api.py` -> Run Flask Backend server.
Whenever, we are going to make an api call to backend application, we are going to use this file.

`crew.py` -> Create crews by connecting Agents and Tasks.

`tasks.py` -> Defining specific task that need to be execute.

`agents.py` -> Store the logic for running all the agents, describing the agents (goal, backstory, etc)

`models.py` -> Setup Pydantic models.
Whenever, we need to start working on the frontend code, we need our data to be formatted in a very specific manner.

`job_manager.py` -> Handle Log CrewAI job status.

`tools\youtube_search_tools.py` -> Search YouTube videos that relate to the companies and job positions.

`pyproject.toml` -> Handle dependencies, project settings and metadata.
Copy and paste the code for pyproject.toml file.
