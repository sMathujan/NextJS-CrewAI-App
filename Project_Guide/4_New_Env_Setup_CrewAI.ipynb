So, created the new virtual environment for the project, by implementing the following.

1. Create `requirements.txt` file.




2. Create the conda environmet.
`conda create -n crewai-nextjs python=3.11`




3. Activate the conda environment.
`conda activate crewai-nextjs`




4. 5- Install the requirements.txt.
`pip install -r requirements.txt`
